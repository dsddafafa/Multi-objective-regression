```python
from scipy.spatial.distance import pdist, squareform
from scipy.stats import pearsonr

def calculate_spearman_corr(X, Y):
    """
    ËÆ°ÁÆóÊñØÁöÆÂ∞îÊõºÁß©Áõ∏ÂÖ≥Á≥ªÊï∞ (Spearman's rank correlation coefficient)
    
    ÂèÇÊï∞:
    X: ÁâπÂæÅÁü©ÈòµÊàñÈ¢ÑÊµãÂÄº (‰∏ÄÁª¥Êï∞ÁªÑÊàñÂàóÂêëÈáè)
    Y: ÁõÆÊ†áÂÄºÂêëÈáè (‰∏ÄÁª¥Êï∞ÁªÑÊàñÂàóÂêëÈáè)
    
    ËøîÂõû:
    rho: ÊñØÁöÆÂ∞îÊõºÁõ∏ÂÖ≥Á≥ªÊï∞ (ËåÉÂõ¥: -1 Âà∞ 1)
    """
    # Á°Æ‰øùËæìÂÖ•ÊòØ‰∏ÄÁª¥Êï∞ÁªÑ
    X = X.flatten()
    Y = Y.flatten()
    
    # ËÆ°ÁÆóÊñØÁöÆÂ∞îÊõºÁß©Áõ∏ÂÖ≥Á≥ªÊï∞
    rho, _ = spearmanr(X, Y)
    
    return rho
```
